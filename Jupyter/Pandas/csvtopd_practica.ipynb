# Práctica 08: inportar datos CSV a Python

Vamos a ver cómo importar csv cuando:
- Punto 1: datos separados por ','
- Punto 2: datos separados por ';'
- Punto 3: no hay nombres de columnas


## Punto 0:

```python
# importar pandas
import pandas as pd
```

## Punto 1:

```python
# caso1: datos separados por ',' 
pd.read_csv('datos_rrss1.csv')
```

## Punto 2: 

```python
# caso2: separado por ';'
pd.read_csv('datos_rrss2.csv', sep=';')
```

## Punto 3: 

```python
# caso 3 : no hay nombres de columnas
pd.read_csv('datos_rrss3.csv',
            header=None,
            names=['Nombre', 'Cantidad', 'ES_FBK', 'Año'])
```
