# Punto 0:


```python
# importar pandas
import pandas as pd
```

```python
fbk = ['Facebook', 2449, True, 2006]
twt = ['X', 339, False, 2006]
ig = ['Instagram', 1000, True, 2010]
yt = ['Youtube', 2000, False, 2005]
lkn = ['LinkedIn', 663, False, 2003]
wsp = ['WhatsApp', 1600, True, 2009]
```

```python
lista_rrss = [fbk, twt, ig, yt, lkn, wsp]
```

```python
# Crear Dataframe a partir de listas
df_rrss=pd.DataFrame(lista_rrss, columns=['Nombre', 'Cantidad', 'es_FBK', 'Año'])
df_rrss
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.plot(x,y)
plt.show()
```

## Punto1:

```python
# Gráfico de líneas utilizando matplotlib
plt.plot(df_rrss['Nombre'], df_rrss['Cantidad'])
plt.show()
```

```python
# Gráfico de puntos utilizando matplotlib
plt.scatter(df_rrss['Nombre'], df_rrss['Cantidad'])
plt.show()
```

```python
# EXTRA
# Hacer lo mismo con métodos de un dataframe de pandas
df_rrss.plot(kind='scatter', x='Nombre', y='Cantidad')
plt.show()
```

## Punto 2

```python
# gráfico de Barras
```

```python
plt.bar(df_rrss['Nombre'], df_rrss['Cantidad'])
plt.show()
```

```python
df_rrss_sort = df_rrss.sort_values('Cantidad', ascending=False)
plt.bar(df_rrss_sort['Nombre'], df_rrss_sort['Cantidad'])
plt.show()
```

```python
# gráfico de barra ordenado y con colores
plt.bar(df_rrss_sort['Nombre'], df_rrss_sort['Cantidad'],
        color=['b', 'r', 'g', 'm', 'k', 'c'])
plt.show()

# b: blue.
# g: green.
# r: red.
# c: cyan.
# m: magenta.
# y: yellow.
# k: black.
# w: white.
```

## Punto 3: 

```python
# Gráficas de tarta/pie
plt.pie(df_rrss['Cantidad'], labels=df_rrss['Nombre'])
plt. show()
```

```python
lista_col_rrss = ['#3b5998', '#FF0000', '#25d366', '#8a3ab9', '#0e76a8', '#000000']
```

```python
plt.pie(df_rrss_sort['Cantidad'],
        labels=df_rrss_sort['Nombre'],
        colors=lista_col_rrss)
plt.show()
```
