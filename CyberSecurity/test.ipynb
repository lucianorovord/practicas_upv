```python
df_cibseg.drop(columns=['id','Flow ID','Src IP','Src Port','Dst IP','Dst Port','Protocol','Timestamp','Attempted Category'], inplace=True) 
df_cibseg.to_csv('wednesday_clean.csv', index=False)
```

```python
df_cibseg['Label'] = df_cibseg['Label'].str.strip()
df_cibseg['Label'] = np.where(df_cibseg['Label'] == 'BENIGN', 0, 1)
df_cibseg
```

```python
df_cibseg['Label'].value_counts()
```

# SPLIT

```python
y = df_cibseg['Label']
x = df_cibseg.drop(columns=['Label'])
```

```python
x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size = 0.30)
```

```python
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.50)
```

```python
print(x_train.shape)
print(x_val.shape)
print(x_test.shape)
```

# SCALING

```python
col = list(x_train.columns.values)
print(col)
```

```python
col = list(x_val.columns.values)
print(col)
```

```python
x_train.replace([np.inf, -np.inf], np.nan, inplace=True)
x_val.replace([np.inf, -np.inf], np.nan, inplace=True)
x_test.replace([np.inf, -np.inf], np.nan, inplace=True)



x_train.fillna(x_train.mean(), inplace=True)
x_val.fillna(x_train.mean(), inplace=True)
x_test.fillna(x_train.mean(), inplace=True)

```

```python
scaler = StandardScaler()

x_train[col] = scaler.fit_transform(x_train[col])

x_val[col] = scaler.transform(x_val[col])
x_test[col] = scaler.transform(x_test[col])

print(x_train.mean(axis=0).round(2))
print(x_train.std(axis=0).round(2))
```

# Guardar en varios CSV los datos de Training, Validación y Test

Para que en el pipeline cuando hagamos las fases finales no tenga que pre-procesar los datos todo el tiempo y reduce drásticamente el tiempo de espera

```python
x_train['Label'] = y_train
x_train.to_csv('cs_train.csv', index=False)
```

```python
x_val['Label'] = y_val
x_val.to_csv('cs_val.csv', index=False)
```

```python
x_test['Label'] = y_test
x_test.to_csv('cs_test.csv', index=False)
```
